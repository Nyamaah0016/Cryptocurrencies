# Cryptocurrencies

### Purpose
Objective of this assignment is to analyze cryptocurrencies by using unsupervised ML.The method used in the assignment are:
Preprocessing the database,reducing the data dimension using Principal Component Analysis, clustering cryptocurrencies using K-Means, visualizing classification results with 2D and 3D scatter plots.
